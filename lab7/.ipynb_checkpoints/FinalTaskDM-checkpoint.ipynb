## **Dataset characteristics:**

---



motorbike_ambulance_calls.csv has the following fields:

              - index: record index
              - date : date
              - season : season (1:springer, 2:summer, 3:fall, 4:winter)
              - yr : year (0: 2011, 1:2012)
              - mnth : month ( 1 to 12)
              - hr : hour (0 to 23)
              - holiday : whether day is holiday or not
              - weekday : day of the week
              - workingday : if day is neither weekend nor holiday is 1, otherwise is 0.
              -  weathersit :
                      - 1: Clear, Few clouds, Partly cloudy, Partly cloudy
                      - 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
                      - 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
                      - 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
              - temp : Normalized temperature in Celsius. The values are divided to 41 (max)
              - atemp: Normalized feeling temperature in Celsius. The values are divided to 50 (max)
              - hum: Normalized humidity. The values are divided to 100 (max)
              - windspeed: Normalized wind speed. The values are divided to 67 (max)
              - cnt: count of total ambulance calls

# Anomaly detection

```python
X = data.drop(columns=['cnt_category'])
y = data['cnt_category']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=175)
```

```python
print(X.columns)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.ensemble import IsolationForest

iso_forest = IsolationForest(contamination=0.05)  
iso_forest.fit(X_train)

y_pred = iso_forest.predict(X_test)

indices = X_test[:, 3]

cnt_values = X_test[:, 12]  

plt.figure(figsize=(20, 30))

plt.scatter(indices[y_pred == 1], cnt_values[y_pred == 1], c='blue', label='Normal', alpha=0.6)

plt.scatter(indices[y_pred == -1], cnt_values[y_pred == -1], c='red', label='Anomaly', alpha=0.6)


plt.title("Anomaly Detection with Isolation Forest", fontsize=14)
plt.xlabel("Hour")
plt.ylabel("Count of calls")
plt.legend()
plt.grid(True)





# Show the plot
plt.show()
```

```python
iso_forest = IsolationForest(contamination=0.05)
iso_forest.fit(X_train)

y_pred = iso_forest.predict(X_test)

indices = X_test[:, 0]

cnt_values = X_test[:, 12]  

plt.figure(figsize=(20, 30))
plt.scatter(indices[y_pred == 1], cnt_values[y_pred == 1], c='blue', label='Normal', alpha=0.6)
plt.scatter(indices[y_pred == -1], cnt_values[y_pred == -1], c='red', label='Anomaly', alpha=0.6)

plt.title("Anomaly Detection with Isolation Forest", fontsize=14)
plt.xlabel("Index")
plt.ylabel("Count of calls")
plt.legend()
plt.grid(True)

plt.show()
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(data.drop(columns=['hr', 'cnt']), 
                                                    data[['hr', 'cnt']], test_size=0.2, random_state=42)


iso_forest = IsolationForest(contamination=0.05)
iso_forest.fit(X_train)


y_pred = iso_forest.predict(X_test)


X_test_with_predictions = pd.DataFrame(X_test, columns=data.drop(columns=['hr', 'cnt']).columns)  # Correct column names
X_test_with_predictions['Anomaly'] = y_pred


X_test_with_predictions['hr'] = X_test.index.to_series().apply(lambda idx: data.loc[idx, 'hr'])
X_test_with_predictions['cnt'] = X_test.index.to_series().apply(lambda idx: data.loc[idx, 'cnt'])

anomalies = X_test_with_predictions[X_test_with_predictions['Anomaly'] == -1]

anomaly_report = {
    'Total Observations': len(X_test),
    'Anomalies Detected': len(anomalies),
    'Anomaly Percentage': (len(anomalies) / len(X_test)) * 100,
    'Anomalous Points': anomalies[['hr', 'cnt']]
}

print("Anomaly Report:")
for key, value in anomaly_report.items():
    if isinstance(value, pd.DataFrame):
        print(f"\n{key}:\n{value.head()}\n")
    else:
        print(f"{key}: {value}")
```
